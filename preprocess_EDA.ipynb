```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("diabetes_012_health_indicators_BRFSS2015.csv")
df.head()
```

```python
df.rename(columns={'Diabetes_012':'Diabetes'}, inplace=True)
```

```python
duplicates = df.duplicated().sum()
print("Duplicate count = ", duplicates)
```

```python
df = df.drop_duplicates()

print("Duplicates Removed Successfully")
print("New Shape:", df.shape)
```

```python
#target variable distribution

sns.countplot(x='Diabetes', data=df)

plt.title("Diabetes Distribution")
plt.xlabel("Diabetes")
plt.ylabel("Count")
plt.show()
```

```python
sns.heatmap(df.corr(),cmap='coolwarm',annot=False)

plt.title("Correlation Heatmap")
plt.show()
```

```python
#bmi distribution
#mental wellness indications
#physical health condition


numerical_columns = ['BMI','MentHlth','PhysHlth']

for i in numerical_columns:
    sns.histplot(df[i], kde=True)
    plt.title(f"Distribution of {i}")
    plt.show()
```

```python
sns.boxplot(x='Diabetes',y='BMI',data=df)

plt.title("BMI vs Diabetes")
plt.show()                                                  #this proves that old people are more likely to have diabetes
```

```python
df.to_csv("cleaned_health_data.csv", index=False)
print("Cleaned Dataset Saved Successfully")
```
